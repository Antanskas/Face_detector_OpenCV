# Imports

```python
from PIL import Image
import numpy as np
import matplotlib.patches as patches

from os import listdir
from os.path import isfile, join

import cv2
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline

working_path = "/content/drive/My Drive/Face_detector/Face_detector_cv2-master/"
!ls "/content/drive/My Drive/Face_detector/Face_detector_cv2-master/"
```

# Functions

```python
def plot_detector_and_save(img_array, coordinates, j):
    fig,ax = plt.subplots(1)
    ax.imshow(img_array)
    
    for i in coordinates:
        (x, y, w, h) = i
        rect = patches.Rectangle((x, y), w, h, linewidth=1.5, edgecolor='r', facecolor='none')
    
        ax.add_patch(rect)
    plt.show()
    fig.savefig(working_path + 'outputs/test_{}.jpg'.format(j))
```

# Detect human face

```python
# load face detector
face_cascade = cv2.CascadeClassifier(working_path + 'haarcascades/haarcascade_frontalface_alt.xml')
```

```python
# list images to test
mypath = working_path + 'images'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
# list of images in gray scale
img_list = []
coordinates_list = []

for img in onlyfiles:
    # img into array
    img = np.array(Image.open(mypath + "/{}".format(img)), dtype=np.uint8)
    
    # img into gray scale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # coordinates of the face in an image
    face = face_cascade.detectMultiScale(gray)
    
    # image and coordinates list
    img_list.append(img)
    coordinates_list.append(face)
    
```

# Plot and save

```python
i = 0
for img_array, coordinates in zip(img_list, coordinates_list):
    plot_detector_and_save(img_array, coordinates, i)
    i = i+1
```
