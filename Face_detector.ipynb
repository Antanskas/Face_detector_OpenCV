# Imports

```python
from PIL import Image
import numpy as np
import matplotlib.patches as patches

from os import listdir
from os.path import isfile, join

import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

# Functions

```python
def plot_detector(img_array, coordinates):
    fig,ax = plt.subplots(1)
    ax.imshow(img_array)
    
    for i in coordinates:
        (x, y, w, h) = i
        rect = patches.Rectangle((x, y), w, h, linewidth=1.5, edgecolor='r', facecolor='none')
    
        ax.add_patch(rect)
    plt.show()
```

# Detect human face

```python
# load face detector
face_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_alt.xml')
```

```python
# list images to test
mypath = 'images'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
# list of images in gray scale
img_list = []
coordinates_list = []

for img in onlyfiles:
    # img into array
    img = np.array(Image.open(mypath + "/{}".format(img)), dtype=np.uint8)
    
    # img into gray scale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # coordinates of the face in an image
    face = face_cascade.detectMultiScale(gray)
    
    # image and coordinates list
    img_list.append(img)
    coordinates_list.append(face)
    
```

# Plot

```python
for img_array, coordinates in zip(img_list, coordinates_list):
    plot_detector(img_array, coordinates)
```
